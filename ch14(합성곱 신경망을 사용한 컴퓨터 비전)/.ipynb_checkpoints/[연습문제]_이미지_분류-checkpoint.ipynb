```python
# dataset from https://www.kaggle.com/juniorbueno/neural-networks-homer-and-bart-classification
# non-commercial use. only for study CNN

import numpy as np
import pandas as pd 

import cv2           
import numpy as np
import os
import zipfile
import tensorflow as tf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
path = "/content/drive/MyDrive/deeplearning/ch14/homer_bart_1"
from google.colab import drive
drive.mount('/content/drive')
```

```python
os.listdir(path)
```

```python
width, height = 224 , 224
train_dir = os.path.join(path)

classes = []
images = []
for i in os.listdir(train_dir):
    if i.startswith("b"):
        classes.append(1)
    else:
        classes.append(0)
    
        
    image = cv2.imread(os.path.join(path,i))
    image = cv2.resize(image, (width, height))
    images.append(image)
```

```python
images_array = np.array(images)
classes_array = np.array(classes)
```

```python
class_names = ["Homer","Bart"]
```

```python
plt.figure(figsize=(10,10))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(images_array[i])
    plt.title(class_names[classes[i]])
    plt.axis(False)
```

```python
images_array = images_array/255
```

### 데이터셋 분류 : 데이터셋의 크기가 크지 않기 때문에 validation set을 사용하지 않기로 함.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(images_array, classes_array, test_size = 0.2, random_state = 1)
```

#### Xception 모델을 활용해 모델 학습

```python
base_model = keras.applications.xception.Xception(weights="imagenet",
                                                  include_top=False)
avg = keras.layers.GlobalAveragePooling2D()(base_model.output)
output = keras.layers.Dense(2, activation="softmax")(avg)
model = keras.models.Model(inputs=base_model.input, outputs=output)
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
optimizer = keras.optimizers.SGD(learning_rate=0.2, momentum=0.9, decay=0.01)
model.compile(loss="sparse_categorical_crossentropy", optimizer=optimizer,
              metrics=["accuracy"])
history = model.fit(X_train, y_train,validation_data=(X_test,y_test),
                    epochs=10)
```

#### 모델 평가

```python
prediction = model.predict(X_test)
pred = []
for i in range(len(prediction)):
  pred.append(np.argmax(prediction[i]))
```

```python
plt.figure(figsize=(10,10))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(X_test[i])
    plt.title(f"Predicted {class_names[pred[i]]},Actual {class_names[y_test[i]]}")
    plt.axis(False)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, pred))
```

```python
sns.heatmap(confusion_matrix(y_test,pred), annot = True)
```
